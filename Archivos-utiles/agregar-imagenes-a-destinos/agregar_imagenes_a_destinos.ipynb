# Agregar imagenes a destinos

```python
import pandas as pd
import requests
import json
```

```python
df = pd.read_json('imagenes_destinos.json', orient='records')
```

```python
with open("imagenes_destinos.json", "r", encoding="utf-8") as f:
    data = json.load(f)

df_imagenes = pd.DataFrame(data)
df_imagenes
```

```python
df_destinos = pd.read_csv('../Viajes/DestinosConAeropuertos/destinos_tiene_aeropuerto_column.csv')
df_destinos
```

```python
df_destinos['img_path'] = df_imagenes['url']
df_destinos
```

```python
df_destinos.to_csv('destinos_con_imagenes_correctas.csv', index=False)
```

```python
url = "https://cdn.sanity.io/images/esqfj3od/production/c1c9069644295a989eaa58f5f18fa124eb438379-1080x720.webp?w=1080&q=65&fit=clip&auto=format"
response = requests.get(url)

if response.status_code == 200:
    with open("image.jpg", "wb") as f:
        f.write(response.content)
    print("Image downloaded successfully!")
else:
    print("Failed to download image.")
```

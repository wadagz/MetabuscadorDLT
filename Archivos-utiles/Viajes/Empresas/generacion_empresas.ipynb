# Generación de empresas

```python
import pandas as pd
```

```python
aerolineas = {
    "nombre": [
        "Aeroméxico",
        "Volaris",
        "Viva Aerobus",
        "Interjet",
        "Aeromar",
    ],
    "descripcion": [
        "La aerolínea nacional más grande, con vuelos nacionales e internacionales.",
        "Aerolínea de bajo costo con amplia cobertura nacional.",
        "Otra aerolínea de bajo costo que opera rutas nacionales e internacionales.",
        "Ofrece vuelos nacionales e internacionales, aunque ha enfrentado dificultades financieras.",
        "Aerolínea regional que conecta ciudades pequeñas y medianas en México.",
    ]
}
aerolineas['tipo'] = ['Aerolínea' for _ in range(len(aerolineas['nombre']))]


autobuses = {
    "nombre": [
        "ADO",
        "Estrella Blanca",
        "Futura",
        "Omnibus de México",
        "Transportes del Norte",
    ],
    "descripcion": [
        "Líder en transporte de pasajeros en México, con rutas nacionales e internacionales.",
        "Ofrece servicios de autobuses de lujo y económicos a diversas ciudades.",
        "Parte del Grupo ADO, especializada en viajes de lujo y turismo.",
        "Conectando múltiples destinos en el país con servicios de calidad.",
        "Proporciona servicios de transporte en diversas rutas nacionales.",
    ]
}
autobuses['tipo'] = ['Autobuses' for i in range(len(autobuses['nombre']))]
```

```python
df_aerolineas = pd.DataFrame(aerolineas)
df_autobuses = pd.DataFrame(autobuses)

print(df_aerolineas)
print(df_autobuses)
```

```python
df_empresas_transporte = pd.concat([df_aerolineas, df_autobuses], ignore_index=True)
df_empresas_transporte['id'] = [i+1 for i in range(len(df_empresas_transporte))]
df_empresas_transporte
```

```python
df_empresas_transporte.to_csv('./empresas_transporte.csv', index=False)
```

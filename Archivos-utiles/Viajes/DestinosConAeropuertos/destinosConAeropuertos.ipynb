```python

def normalizar(texto):
    texto = texto.lower()
    texto = unicodedata.normalize('NFKD', texto)
    return ''.join(c for c in texto if not unicodedata.combining(c))

def string_en_lista(cadena, lista):
    cadena_norm = normalizar(cadena)
    return any(normalizar(item) == cadena_norm for item in lista)
```

```python
import pandas as pd
```

```python
df_destinos = pd.read_csv('info_destinos.csv')
df_municipios = pd.read_csv('municipios.csv')
df_estados = pd.read_csv('estados.csv')
```

## Remover registros vacios de destinos

```python
df_destinos.dropna(axis='index', how='any', inplace=True)
```

## Hace cross reference para obtener el id del estado correspondiente a cada destino.

```python
def remove_estado_string_from_estado_field(row):
    if 'Estado de' in row['estado']:
        splitted = row['estado'].split()
        estado_name = ""
        for i, word in enumerate(splitted[2:]):
            if i == 0: estado_name += word
            else: estado_name += " " + word
        return estado_name
    return row['estado']

df_destinos['estado'] = df_destinos.apply(remove_estado_string_from_estado_field, axis='columns')
```

```python
def get_estado_id_for_destino(row):
    for _, estado_row in df_estados.iterrows():
        if estado_row['nom_ent'] == row['estado']: return estado_row['cod_ent']
    return 0

sr_id_estados_para_destino = df_destinos.apply(get_estado_id_for_destino, axis='columns')

df_destinos['id_estado'] = sr_id_estados_para_destino
```

## Hace cross reference para obtener el id del municipio correspondiente a cada destino.

```python
def remove_municipio_string_from_municipio_field(row):
    if 'Municipio de' in row['municipio']:
        splitted = row['municipio'].split()
        municipio_name = ""
        for i, word in enumerate(splitted[2:]):
            if i == 0: municipio_name += word
            else: municipio_name += " " + word
        return municipio_name
    return row['municipio']

df_destinos['municipio'] = df_destinos.apply(remove_municipio_string_from_municipio_field, axis='columns')
```

```python
def get_municipio_id_for_destino(row):
    for _, municipio_row in df_municipios.iterrows():
        if municipio_row['cod_ent'] == row['id_estado'] and municipio_row['nom_mun'] == row['municipio']:
            return municipio_row['cod_mun']
    return 0

df_destinos['id_municipio'] = df_destinos.apply(get_municipio_id_for_destino, axis='columns')
```

```python
df_destinos.to_csv('destinos.csv', index=False)
```

## Hoteles obtenidos de datosgob_2024.csv

### Carga y filtrado de registros por hoteles

```python
df = pd.read_csv('./datosgob_2024.csv', encoding='latin-1')
```

```python
df_hotels = df[df.Giro.str.contains('hotel', case=False, na=False, regex=True)]
df_hotels
```

```python
df_hotels = df_hotels.drop(columns = ['SCIAN', 'Rango de Empleados', 'Registrado por', 'Giro'])
df_hotels = df_hotels.rename(columns={ 'C.P.': 'CP' })
df_hotels
```

```python
df_hotels['tipo_hospedaje'] = ['Hotel' for _ in range(len(df_hotels))]
df_hotels['id_propietario'] = [x+1 for x in range(len(df_hotels))]
df_hotels
```

### Generación de Propietarios a partir de los datos de correo y teléfono

```python
df_propietarios = pd.DataFrame(df_hotels.filter(['Teléfono', 'E-mail']))
df_propietarios.fillna('Desconocido', axis='columns', inplace=True)

def get_name_from_email(row):
    return row['E-mail'].rsplit('@')[0]

df_propietarios['nombre'] = df_propietarios.apply(get_name_from_email, axis='columns')

df_propietarios
```

### Direcciones de hoteles

```python
import re

def limpiar_calle(domicilio):
    # Elimina contenido tipo "int X", "int S/N", "int N/A"
    domicilio = re.sub(r'\bint\b\s*[\w\./-]+', '', domicilio, flags=re.IGNORECASE)

    # Elimina "KM", "KM.", "KILÓMETRO", con número o decimal
    domicilio = re.sub(r'\bK(M|IL[ÓO]METRO)?\.?\s*\d+(\.\d+)?', '', domicilio, flags=re.IGNORECASE)

    # Elimina número exterior si está justo después del nombre de la calle
    domicilio = re.sub(r'\s+\d+\b', '', domicilio)

    # Elimina múltiples espacios
    domicilio = re.sub(r'\s{2,}', ' ', domicilio)

    # Quita espacios iniciales/finales
    return domicilio.strip()

# Remueve información innecesario del domicilio.
def limpiar_calle_v2(domicilio):
    domicilio = domicilio.strip()

    # Caso especial: si es "Calle [número] [otro número]"
    match = re.match(r'^(Calle\s+\d+)\b(.*)', domicilio, re.IGNORECASE)
    if match:
        return match.group(1)

    # Quitar partes de "int X", "int S/N", etc.
    domicilio = re.sub(r'\bint\b\s*[\w\./-]+', '', domicilio, flags=re.IGNORECASE)

    # Quitar "KM", "KM.", "KILÓMETRO" con número o decimal
    domicilio = re.sub(r'\bK(M|IL[ÓO]METRO)?\.?\s*\d+(\.\d+)?', '', domicilio, flags=re.IGNORECASE)

    # Quitar número exterior (ej. "405", "1234") si no es parte del nombre
    domicilio = re.sub(r'\s+\d+\b', '', domicilio)

    # Quitar múltiples espacios
    domicilio = re.sub(r'\s{2,}', ' ', domicilio)

    return domicilio.strip()

```

```python
df_hotels['Domicilio'] = df_hotels['Domicilio'].apply(limpiar_calle_v2)
df_hotels['Domicilio']
```

```python
import requests
from faker import Faker
import time

Faker.seed(42)
fake = Faker("es_MX")

def buscar_direccion_nominatim(calle, ciudad, estado, pais="Mexico", espera=1):
    """
    Consulta la API de Nominatim con datos de dirección y devuelve la respuesta JSON.
    
    Parámetros:
        calle (str): Nombre de la calle (sin número interior).
        ciudad (str): Nombre del municipio o ciudad.
        estado (str): Estado de la República.
        pais (str): País (por defecto 'Mexico').
        espera (int): Tiempo de espera entre solicitudes (segundos).
        
    Retorna:
        dict | None: Primer resultado de la búsqueda o None si no hay resultados.
    """
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "amenity": 'Hotel cerca', # Obtiene hoteles cercanos a la dirección provista.
        "street": calle,
        "city": ciudad,
        "state": estado,
        "country": pais,
        "format": "jsonv2",
        "addressdetails": 1,
        "extratags": 1,
    }

    headers = {
        'User-Agent': 'metabuscador-dtl',  # Si no se pone un user-agent se enoja la API
        'Accept': 'application/json',  # Para aceptar respuesta en JSON
    }

    try:
        response = requests.get(url, params=params, headers=headers)
        response.raise_for_status()
        resultados = response.json()

        if resultados:
            print(resultados)
            return resultados
        else:
            print('Sin resultados')
            return None
    except requests.RequestException as e:
        print(f"Error en la solicitud: {e}")
        return None
    finally:
        time.sleep(espera)  # Para evitar ser bloqueado por exceso de solicitudes

# Agrega columnas de latitud y longitud
def geocodificar_fila(fila):
    resultado = buscar_direccion_nominatim(fila['Domicilio'], fila['Municipio'], fila['Estado'])
    if resultado:
        return pd.Series([resultado.get('lat'), resultado.get('lon')])
    else:
        return pd.Series([None, None])

def generar_hoteles_con_direccion(fila):
    resultados = buscar_direccion_nominatim(fila['Domicilio'], fila['Municipio'], fila['Estado'])
    if resultados is not None:
        for resultado in resultados:
            hotel = dict()
            propietario = dict()
            direccion = dict()

            hotel['nombre'] = 'Hotel ' + fake.company() if resultado.get('name') == '' else re.sub(r"[^\w\s,]", '', resultado.get('name'), flags=re.UNICODE)
            hotel['precio'] = random.randint(1000, 20000)

            propietario['nombre'] = fake.name()
            propietario['telefono'] = fake.phone_number()
            propietario['email'] = fake.email()

            direccion['nombre'] = resultado.get('address').get('road')
            direccion['estado'] = resultado.get('address').get('state')
            direccion['municipio'] = resultado.get('address').get('county')
            direccion['ciudad'] = resultado.get('address').get('city')
            direccion['cp'] = resultado.get('address').get('postcode')
            direccion['lat'] = resultado.get('lat')
            direccion['lon'] = resultado.get('lon')

            hoteles.append(hotel)
            propietarios.append(propietario)
            direcciones.append(direccion)
```

```python
df_hoteles = pd.DataFrame(hoteles)
df_direcciones = pd.DataFrame(direcciones)
df_propietarios = pd.DataFrame(propietarios)
```
